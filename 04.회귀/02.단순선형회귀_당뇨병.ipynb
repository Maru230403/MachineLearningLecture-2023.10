## 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- 당뇨병 데이터 탐색

```python
from sklearn.datasets import load_diabetes
diab = load_diabetes()
df = pd.DataFrame(diab.data, columns=diab.feature_names)
df['target'] = diab.target
df.head()
```

```python
df.describe()   # 분포가 N(0, 0.00476)으로 이미 정규화되어 있음
```

```python
print(diab.DESCR)
```

- 10개의 feauture와 target간의 상관관계도

```python
fig, axs = plt.subplots(2, 5, figsize=(16, 8), sharey=True)
for i, feature in enumerate(diab.feature_names):
    row, col = i // 5, i % 5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- 10개 feature의 R squared 값

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
for feature in diab.feature_names:
    X = df[feature].values.reshape(-1, 1)
    lr.fit(X, diab.target)
    score = lr.score(X, diab.target)
    print(f'{feature}: {score:.4f}')
```

### BMI vs target

```python
# 훈련/테스트 데이터셋 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1, 1), diab.target, test_size=0.2, random_state=2023
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 가중치, 바이어스
lr.coef_, lr.intercept_
```

```python
# R squared
lr.score(X_test, y_test)
```

```python
# 회귀선
xs = np.array([X_train.min(), X_train.max()])
ys = lr.coef_[0] * xs + lr.intercept_
```

- 시각화

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, marker='^', label='test')
plt.plot(xs, ys, c='red', lw=3)
plt.title('BMI vs target', fontsize=15)
plt.xlabel('BMI'), plt.ylabel('target'), plt.legend()
plt.show()
```

### s3(HDL) vs target

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df.s3.values.reshape(-1, 1), diab.target, test_size=0.2, random_state=2023
)
lr2 = LinearRegression()
lr2.fit(X_train2, y_train2)
lr2.score(X_test2, y_test2)
```

```python
xs2 = np.array([X_train2.min(), X_train2.max()])
ys2 = lr2.coef_[0] * xs2 + lr2.intercept_
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train2, y_train2, label='train')
plt.scatter(X_test2, y_test2, marker='^', label='test')
plt.plot(xs2, ys2, c='red', lw=3)
plt.title('s3(HDL) vs target', fontsize=15)
plt.xlabel('s3(HDL)'), plt.ylabel('target'), plt.legend()
plt.show()
```
